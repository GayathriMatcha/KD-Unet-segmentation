```python
def dice_coef(inputs, targets):
    y_inputs = inputs.flatten()
    y_targets = targets.flatten()
#     intersection = np.sum(y_true_f * y_pred_f)
    intersection = (y_inputs * y_targets).sum()
    smooth = 0.0001
    dice = (2. * intersection + smooth) / (y_inputs.sum() + y_targets.sum() + smooth)
#     intersection = (inputs * targets).sum()                            
#     dice = (2.*intersection + smooth)/(inputs.sum() + targets.sum() + smooth)  

    return dice

def dice_coef_multilabel(y_true, y_pred):
    dice=0
    numLabels=4
    for index in range(numLabels):
        dice += dice_coef(y_true[index,:,:], y_pred[index,:,:])
    return dice/numLabels # taking average


link=
```

```python
def hd_distance(x, y):

    if np.count_nonzero(x) == 0 or np.count_nonzero(y) == 0:
        return np.array([np.Inf])

    indexes = np.nonzero(x)
    distances = edt(np.logical_not(y))

    return np.array(np.max(distances[indexes]))

def compute(pred,target):
    assert (
        pred.shape[1] == 1 and target.shape[1] == 1
    ), "Only binary channel supported"

    pred = (pred > 0.5).byte()
    target = (target > 0.5).byte()

    right_hd = torch.from_numpy(
        self.hd_distance(pred.cpu().numpy(), target.cpu().numpy())
    ).float()

    left_hd = torch.from_numpy(
        self.hd_distance(target.cpu().numpy(), pred.cpu().numpy())
    ).float()

    return torch.max(right_hd, left_hd)
def hd_multilabel(y_true, y_pred):
    dice=0
    numLabels=4
    for index in range(numLabels):
        dice += compute(y_true[index,:,:], y_pred[index,:,:])
    return dice/numLabels # taking average
```
