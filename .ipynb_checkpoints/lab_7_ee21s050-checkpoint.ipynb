```python
# Q.1
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,20,0.01)
y = np.arange(0,20,0.01)
X,Y = np.meshgrid(x,y)

Z = (X - 10)**2 + (Y - 10)**2
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(X,Y,Z)
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('J(w1,w2)')
```

```python
plt.contour(X,Y,Z)
plt.xlabel('w1')
plt.ylabel('w2')
plt.title('J(w1,w2)')
```

```python
# Q.2
x = np.arange(2,4,0.01)
J = x**2 + 54/x
plt.plot(x,J)
plt.xlabel('w')
plt.ylabel('J(w)')
```

```python
# Bracketing method
def J(x):
    return x**2 + 54/x

a=2
b=4
n=10
del_w = (b-a)/n
w1 = a
w2 = w1 + del_w
w3 = w2 + del_w

while J(w1)>=J(w2) and J(w2)>=J(w3):
    w1 = w2
    w2 = w3
    w3 = w2 + del_w
    
print(w1,w2,w3)
```

```python
# Interval halving method
a = w1
b = w3
w_m = (a+b)/2
L = b - a
w1 = a + L/4
w2 = b - L/4

while abs(L) > 10**(-2):
    if J(w1) < J(w_m):
        b = w_m
        w_m = w1
        L = b - a
        w1 = a + L/4
        w2 = b - L/4
        
    else:
        if J(w2) < J(w_m):
            a = w_m
            w_m = w2
            L = b - a
            w1 = a + L/4
            w2 = b - L/4
        
        else:
            a = w1
            b = w2
            L = b - a
            w_m = (a+b)/2
            w1 = a + L/4
            w2 = b - L/4

print(a,b)
```

```python
# Newton Raphson method
def del_J(x):
    return 2*x - 54/x**2

def del2_J(x):
    return 2 + (2*54)/x**3

w_1 = a
w_2 = w_1 - del_J(w_1)/del2_J(w_1)

while abs(del_J(w_2)) > 10**(-3):
    w_1 = w_2
    w_2 = w_1 - del_J(w_1)/del2_J(w_1)
    
print(w_1, w_2)
```

```python
# Q.3
x = np.arange(0,np.pi,0.01)
J = np.sin(x)
plt.plot(x,J)
plt.xlabel('w')
plt.ylabel('J(w)')
```

```python
# Bracketing method
def J(x):
    return -np.sin(x)

a=0
b=np.pi
n=10
del_w = (b-a)/n
w1 = a
w2 = w1 + del_w
w3 = w2 + del_w

while J(w1)>=J(w2) and J(w2)>=J(w3):
    w1 = w2
    w2 = w3
    w3 = w2 + del_w
    
print(w1,w2,w3)
```

```python
# Interval halving method
a = w1
b = w3
w_m = (a+b)/2
L = b - a
w1 = a + L/4
w2 = b - L/4

while abs(L) > 10**(-2):
    if J(w1) < J(w_m):
        b = w_m
        w_m = w1
        L = b - a
        w1 = a + L/4
        w2 = b - L/4
        
    else:
        if J(w2) < J(w_m):
            a = w_m
            w_m = w2
            L = b - a
            w1 = a + L/4
            w2 = b - L/4
        
        else:
            a = w1
            b = w2
            L = b - a
            w_m = (a+b)/2
            w1 = a + L/4
            w2 = b - L/4

print(a,b)
```

```python
# Newton Raphson method
def del_J(x):
    return -np.cos(x)

def del2_J(x):
    return np.sin(x)

w_1 = a
w_2 = w_1 - del_J(w_1)/del2_J(w_1)

while abs(del_J(w_2)) > 10**(-4):
    w_1 = w_2
    w_2 = w_1 - del_J(w_1)/del2_J(w_1)
    
print(w_1, w_2)
```
