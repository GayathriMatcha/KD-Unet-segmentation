```python
#Q.2
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,20,0.01)
y = np.arange(0,20,0.01)
X,Y = np.meshgrid(x,y)

Z = (X - 10)**2 + (Y - 10)**2
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(X,Y,Z)
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('J(w1,w2)')
plt.show()
```

```python
plt.contour(X,Y,Z)
plt.xlabel('w1')
plt.ylabel('w2')
plt.title('J(w1,w2)')
plt.show()
```

```python
#Fixed alpha
def J(x, y):
    return (x - 10)**2 + (y - 10)**2

def grad(x, y):
    return (2*(x-10), 2*(y-10))

x,y = 2, 5
alpha = 0.1
i = 0

x_step = []
y_step = []
fig = plt.figure(figsize=(12,9))
ax = plt.axes()
ax.set_title('J(w1,w2)')
ax.plot(x,y,marker="o")

while abs(J(x,y)) > 10**(-3):
    plt.plot(x,y,marker='*')
    print(i, x, y)
    x_step.append(x)
    y_step.append(y)
    x = x - alpha * grad(x,y)[0]
    y = y - alpha * grad(x,y)[1]
    i += 1

ax.plot(x,y,marker="X")
plt.plot(x_step,y_step)
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_xlabel('w1')
ax.set_ylabel('w2')
print('final: ', x, y)
```

```python
#Variable alpha
def J(x, y):
    return (x - 10)**2 + (y - 10)**2

def grad(x, y):
    return (2*(x-10), 2*(y-10))

def alpha(x,y):
    kx, ky = grad(x,y)
    return (x + y - 20)/(kx + ky)

def inter_half(a,b,w_1,w_2,del_w_1, del_w_2):
    
    w_m = (a+b)/2
    L = b - a
    w1 = a + L/4
    w2 = b - L/4

    while abs(L) > 10**(-2):
        if J(w_1 - w1 * del_w_1, w_2 - w1 * del_w_2) < J(w_1 - w_m * del_w_1, w_2 - w_m * del_w_2):
            b = w_m
            w_m = w1
            L = b - a
            w1 = a + L/4
            w2 = b - L/4

        else:
            if J(w_1 - w2 * del_w_1, w_2 - w2 * del_w_2) < J(w_1 - w_m * del_w_1, w_2 - w_m * del_w_2):
                a = w_m
                w_m = w2
                L = b - a
                w1 = a + L/4
                w2 = b - L/4

            else:
                a = w1
                b = w2
                L = b - a
                w_m = (a+b)/2
                w1 = a + L/4
                w2 = b - L/4
                
    return w1

x,y = 2, 5
i = 0
a, b = 5, -5
x_step = []
y_step = []
x_step.append(x)
y_step.append(y)
fig = plt.figure(figsize=(12,9))
ax = plt.axes()
ax.set_title('J(w1,w2)')
ax.plot(x,y,marker="o")

while abs(J(x,y)) > 10**(-3):
    plt.plot(x,y,marker='*')
    print(i, x, y)
    x = x - inter_half(a,b,x,y,grad(x,y)[0],grad(x,y)[1]) * grad(x,y)[0]
    y = y - inter_half(a,b,x,y,grad(x,y)[0],grad(x,y)[1]) * grad(x,y)[1]
    x_step.append(x)
    y_step.append(y)
    i += 1

ax.plot(x,y,marker="X")
plt.plot(x_step,y_step)
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_xlabel('w1')
ax.set_ylabel('w2')
print('final: ', x, y)
```
